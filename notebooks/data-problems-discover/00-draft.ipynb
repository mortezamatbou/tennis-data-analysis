##### Game info rows for signle match event contain detail of games and sets

| column            | description                                                   |
| :---              | :---                                                          |
| set_id            | number of current set                                         |
| game_id           | number of game in current set                                 |
| point_id          | order of serve                                                |
| home_point        | home point in current game. values 0 15 30 40 A(Advantage)    |
| away_point        | away point in current game. values 0 15 30 40 A(Advantage)    |
| point_description | ---------------------------                                   |
| home_point_type   | type of point 1=loss 5=take                                   |
| away_point_type   | type of point 1=loss 5=take                                   |
| home_score        | total game point in current set for home player               |
| away_score        | total game point in current set for away player               |
| serving           | which player serving in current game. 1=home 2=away           |
| scoring           | which player win current game. 1=home 2=away                  |
